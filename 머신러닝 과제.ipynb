```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
X = np.sin(np.linspace(0,math.pi,10))
Y = np.sin(np.linspace(0,math.pi,10))
Z = np.zeros((10,10))

for i,x in enumerate(X):
    for j,y in enumerate(Y):
        Z[i,j]=(X[i]+Y[j])/2
        
plt.imshow(Z)
plt.colorbar()
```

```python
data = np.sin(np.linspace(0,math.pi,10000)).reshape(100,100)
plt.imshow(data.T)
plt.colorbar()
```

```python
data1 = np.random.randn(10000)
data2 = np.random.randn(10000)
plt.scatter(data1,data2,alpha=0.01)
plt.title('Random scatter plot')
```

```python
fig, ax = plt.subplots(2,2)
grid = plt.GridSpec(2,2,wspace=0.5, hspace=0.4)

f1 = np.random.normal(loc=0, scale=1, size=10000)
f2 = np.random.normal(loc=3, scale=.5, size=10000)


plt.subplot(grid[0,0]).hist(f1, bins=200, color='red', alpha=.7)
plt.title('Red histogram')
plt.subplot(grid[0,1]).hist(f2, bins=200, color='blue', alpha=.7)
plt.title('blue histogram')

plt.subplot(grid[1,0:]).hist(f1, bins=200, color='red',alpha=.7,label='loc = 0, scale = 1')
plt.subplot(grid[1,0:]).hist(f2, bins=200, color='blue', alpha=.5, label='loc = 3, scale = .5')
plt.legend()
```
