```python
xorg = ddf.values 
xorg
```

```python
# missing value detection
naidx = np.isnan(xorg) # NaN 여부 확인 (True, False)
naidx
```

```python
naidx.sum() # NaN 갯수 (True는 1로, False는 0로)
```

```python
ddf.info()
```

```python
naidx.sum(axis=0) # 행방향 NaN index 갯수
```

```python
naidx.sum(axis=1) # 열방향 NaN index 갯수
```

```python
xorg[naidx] # NaN 값 확인
```

```python
# simple missing value imputation
from sklearn.impute import SimpleImputer
```

```python
f = SimpleImputer(missing_values=np.nan,strategy='mean')
xfill = f.fit_transform(xorg)
np.isnan(xfill).sum() # NaN index 갯수=0
xfill[naidx] # 기존 NaN 위치에 채워진 값 확인
```

```python
f = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0)
xfill = f.fit_transform(xorg) # NaN index 갯수=0
xfill[naidx]  # 기존 NaN 위치에 채워진 값 확인
```
